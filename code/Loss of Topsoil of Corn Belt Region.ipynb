# About the Dataset: Estimates of topsoil loss of Corn Belt region.

Topsoil is the uppermost layer of soil that provides the most nutrients for the crop. Reduced topsoil layer will definitely affects the yield of the crop. Thus, resulting in economic loss too.

This dataset includes information at count- and state-level across the Corn Belt region. 

"Topsoil loss was estimated from the relationship between subsoil exposure and topography and averaged across each site."
"The relationship between topsoil loss and topographic curvature was used to up-scale and predict topsoil and economic losses at the county and state-levels across the entire 375,000 km2 study area."

Temporal Coverage: Non-continuous satellite imagery captured from April 2003 to July 2018

Citation:
Thaler, E., I. Larsen, and Q. Yu. 2020. Remote Sensing Derived Topsoil and Agricultural Economic Losses, Midwestern USA. ORNL DAAC, Oak Ridge, Tennessee, USA. https://doi.org/10.3334/ORNLDAAC/1774

A full data dictionary is as listed below:

- county_name - county name
- state_name - state name
- area_ha (ha) - land area in hectares
- farm_cnt - number of farms
- farm_area_mean_ac (ac) - mean farm area in acres
- farm_area_km2 (km2)	- farmland area in square kilometers
- corn_planted_ac (ac) - corn planted in acres
- soy_planted_ac	 (ac) - soybean planted in acres
- corn_harvest_ac (ac) - corn harvest in acres
- corn_harvest_ha (ha) - corn harvest in hectares
- soy_harvest_ac	 (ac) - soybean harvest in acres
- soy_harvest_ha	 (ha) - soybean harvest in hectares
- corn_yield_kg_per_ha (kg/ha) - corn yield in kilograms per hectare
- soy_yield_kg_per_ha (kg/ha) - soybean yield in kilograms per hectare
- corn_econ_loss_usd (usd) - economic loss from decreased corn yield in United States dollars
- soy_econ_loss_usd (usd) - economic loss from decreased soybean yield in United States dollars
- farm_econ_loss_mean_usd (usd) - mean economic loss per farm in United States dollars
- econ_loss_mean_usd (usd) - mean economic loss in United States dollars
- econ_loss_usd_per_ha	(usd/ha) - mean economic loss in United States dollar per hectare
- econ_loss_usd_per_km2 (usd/km2) - mean economic loss in United States dollar per square kilometers
- topsoil_loss_km2 (km2) - topsoil loss in square kilometers
- state_econ_loss_usd (usd) - mean economic loss for state in United States dollars
- corn_planted_ha – corn planted in hectares
- soy_planted_ha – soy planted in hectares
- topsoil_loss_ha – topsoil loss in hectares

# Creating new columns:

1) Corn and soybean planted area in unit of hectare with variable name of 'corn_planted_ha' and 'soy_planted_ha' respectively.
##### Note: The conversion of unit ha to ac in this data is by the factor of 2.469135802469136

2) Topsoil loss in unit of hectare with variable name of 'topsoil_loss_ha'.
##### Note: The conversion of unit km2 to ha in this data is by the factior of 100



```python
#Creating 2 new column for corn_planted_ha and soy_planted_ha and 1 column for topsoil_loss_ha

topsoil['corn_planted_ha'] = topsoil['corn_planted_ac'].apply(lambda x: x/2.469135802469136)
topsoil['soy_planted_ha'] = topsoil['soy_planted_ac'].apply(lambda x: x/2.469135802469136)
topsoil['topsoil_loss_ha'] = topsoil['topsoil_loss_km2'].apply(lambda x: x*100)
topsoil.head()
```

##### 7. Build Data_Analyzer class (mainly to analyze the dataset).

```python
class Data_analyzer:
    
    
    def __init__(self):
        print("Hi. Want to analyze something?")
    
    #return the statistical analysis of topsoil loss of agricultural land across Cornbelt Region.
    def stats_analysis(self, var1):
        print('\n')
        self.var1 = var1
        if self.var1 == 'soil loss':
            print('This is statistical analysis of topsoil loss of Cornbelt Region.')
            print(topsoil['topsoil_loss_ha'].describe())
        
    #return the sum of desired data variables (farm area, corn yield/ soy yield/ economy loss)              
    def total(self, var2):
        print('\n')
        self.var2 = var2
        if self.var2 == 'farm area':
            print('The total area of farm (ha) for each states are:')
            print(topsoil.groupby(['state_name'])[['farm_area_ha']].sum())
        elif self.var2 == 'corn yield':
            print('This is total yield of corn (kg/ha) for each states.')
            print(topsoil.groupby(['state_name'])[['corn_yield_kg_per_ha']].sum())
        elif self.var2 == 'soy yield':
            print('This is total yield of soy (kg/ha) for each states.')
            print(topsoil.groupby(['state_name'])[['soy_yield_kg_per_ha']].sum())
        elif self.var2 == 'economy loss':
            print('This is total mean of economic loss for each states.')
            print(topsoil.groupby(['state_name'])[['econ_loss_mean_usd']].sum())
    
    #Return the highest value of loss for topsoil (soil) and economy (economy).
    def max_loss(self, var3):
        print('\n')
        self.var3 = var3
        if self.var3 == 'soil':
            print('The maximum loss for counties of each states area: ')
            print(topsoil.groupby(['state_name'])[['county_name', 'topsoil_loss_ha']].max())
        elif self.var3 == 'economy':
            print('The max of mean economic loss (usd) for counties of each states are: ')
            print(topsoil.groupby(['state_name'])[['county_name', 'econ_loss_mean_usd']].max())
    
    # We assumed that when the amount of topsoil loss is more than 60000 ha, the situation is considered bad.
    # so, eturn the counties and states name that suffered topsoil loss more than 60000 ha.
    def take_action(self):
        print ('\n')
        print('The amounts of topsoil loss for following agricultural lands had exceeded 60000 ha, we must take action now!')
        print(topsoil[(topsoil['topsoil_loss_ha'] > 60000)][['state_name', 'county_name']])
        
```

```python
analyze_topsoil = Data_analyzer()
analyze_topsoil.stats_analysis('soil loss')
analyze_topsoil.total('soy yield')
analyze_topsoil.max_loss('economy')
analyze_topsoil.take_action()
```
